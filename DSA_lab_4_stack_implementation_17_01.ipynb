```python
class Node:
  def __init__(self,data):
    self.data=data
    self.next=None
class stack:
  def __init__(self):
    self.top=None
  def push(self):
    x=int(input("enter element to be inserted into stack :"))
    new_node=Node(x)
    if self.top is None:
      self.top =new_node
      self.top.next=None
    else:
      new_node.next=self.top
      self.top=new_node
  def pop(self):
    if self.top is None:
      print("Stack is empty")
    elif self.top.next is None:
      print("popped element is :",self.top.data)
      print("--------")
      self.top.next=None
    else:
      print("popped element is :",self.top.data)
      print("--------")
      temp=self.top
      self.top=temp.next
      temp.next=None
  def display(self):
    if self.top is None:
      print("stack is empty")
      print("------")
    else:
      print("elements of stack are:")
      temp=self.top
      while temp:
        print(temp.data)
        temp=temp.next
      print("Top element of stack is :",self.top.data)
      print("-------")
s=stack()
while (1):
  print("enter the option from below")
  print("1. push \n 2. pop operation \n 3. display")
  option=int(input())
  if option==1:
    print("push  operation")
    print("------")
    s.push()
  elif option==2:
    print("pop operation")
    print("------")
    s.pop()
  elif option==3:
    print("display operation")
    print("-------")
    s.display()
  else:
    break



```
